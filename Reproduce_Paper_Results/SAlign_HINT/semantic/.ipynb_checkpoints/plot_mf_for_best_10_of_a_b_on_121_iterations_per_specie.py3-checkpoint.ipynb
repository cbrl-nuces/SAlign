```python
import matplotlib.pyplot as plt
import numpy as np

file = '/Users/imran/Documents/semantic/output/missed/salign.txt'
rs = open(file, 'r').readlines()

def plot_graph(start, end, term):
    first_121_rows = rs[start: end]
    maxs = []
#     print(first_121_rows[0],'\n', first_121_rows[120])

    sp = []

    if term == 'mf':
        label_x = "Molecular Function"
        index_c = 2
        color = 'r'
    elif term == 'bp':
        label_x = 'Biological Process'
        color = 'b'
        index_c = 3
    else:
        color = 'c'
        label_x = 'Cellular Component'
        index_c = 4

    for r in first_121_rows:
        col = r.split()
        sp.append(float(col[index_c]))
    temp = []
    print(sp)
#     temp = sp.copy()
    temp = sp[:]
    for i in range(1,10):
        tmp = max(temp)
        max_index = temp.index(tmp)
        maxs.append([max_index, tmp])
        temp[max_index] = 0

    label_line = ''
    label_line = 'max \t\t iteration \t alpha \t beta \t sequence \t structure \n'
    for mx in maxs:
        max_val_row = ''
        iterr = mx[0]
        max_val_row = first_121_rows[iterr].split()

        if iterr < 10:
            iterr = "0{iterr}".format(iterr=iterr)
        label_line = label_line + '{maxx} \t\t {iterr}\n'.format(maxx=mx[1], iterr=iterr)

    title='Mouse-Yeast'
    plt.figure( figsize=(50, 15))    
    plt.xticks(np.arange(0,122,1))
    plt.plot(sp,'{color}^'.format(color=color), linestyle='solid', label=label_line.expandtabs())
    plt.title(title, fontsize=50)
    plt.xlabel('Iterations ->', fontsize=30)
    plt.ylabel(label_x, fontsize=30)
    lgnd = plt.legend(prop={'size': 25}, bbox_to_anchor = (0.5, -0.3), loc = 'center')
    plt.savefig("plots/bma/legend_display_individual_pair/max/{term}/{title}_{term}".format(title=title, term=term), bbox_extra_artists=[lgnd], bbox_inches='tight')

for i in range(0, 1):
    if (i == 0):
        prev_end = 0
        i = 1
    start = 1 + prev_end
    end = start + 121
    prev_end = end
    
    plot_graph(start, end, 'bp')
    
```
