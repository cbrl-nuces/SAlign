```python
import matplotlib.pyplot as plt
import numpy as np

def trans(sp):
    if(sp == "MusMusculus"):
        sp1_t = "Mouse"
    if(sp == "CaenorhabditisElegans"):
        sp1_t = "Worm"
    if(sp == "DrosophilaMelanogaster"):
        sp1_t = "Fly"
    if(sp == "HomoSapiens"):
        sp1_t = "Human"
    if(sp == "SaccharomycesCerevisiaeS288C"):
        sp1_t = "Yeast"
    return sp1_t

file = '/Users/imran/Documents/semantic/output/bma.txt'
rs = open(file, 'r').readlines()

first_121_rows = rs[977: 1098]
mf = []
bp = []
cc = []
print(first_121_rows[0],'\n', first_121_rows[120])

    
for r in first_121_rows:
    col = r.split()
    mf.append(float(col[2]))
    bp.append(float(col[3]))
    cc.append(float(col[4]))

    sp1, sp2 = first_121_rows[1].split()[1].split('-')
    sp1 = trans(sp1)
    sp2 = trans(sp2)
    
title= "{sp1}-{sp2}".format(sp1=sp1, sp2=sp2)
plt.figure( figsize=(15, 7))    
plt.xticks(np.arange(0,121,15), fontsize=15)
plt.yticks( fontsize=15)

# plt.plot(cc,'{color}^'.format(color='c'), linestyle='solid', label='Cellular Component')
plt.plot(mf,'{color}^'.format(color='r'), linestyle='solid', label='Molecular Functionality')
plt.plot(bp,'{color}^'.format(color='b'), linestyle='solid', label="Biological Process")

plt.title(title, fontsize=20)
plt.xlabel('Grid search for single pair', fontsize=20)
plt.ylabel("AFS", fontsize=20)
plt.legend(prop={'size': 15})
plt.savefig("plots/bma/{title}_{term}".format(title=title, term='go'))
```
