```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

def required_columns(req_arr, index_c):
    formatted_arr = []
    for req in req_arr:
        col = req.split()
        formatted_arr.append(float(col[index_c]))
    return formatted_arr

term = 'bp'
graph_criteria = 'scores'
folder = "/Users/imran/Documents/semantic/output/v1/ready-to-plot"

hubalign_output = open('{folder}/hubalign.txt'.format(folder=folder),       'r').readlines();
ibnal_output = open('{folder}/ibnal.txt'.format(folder=folder),             'r').readlines();
magnapp_output = open('{folder}/magnapp.txt'.format(folder=folder),         'r').readlines();
modulealign_output = open('{folder}/modulealign.txt'.format(folder=folder), 'r').readlines();
netal_output = open('{folder}/netal.txt'.format(folder=folder),             'r').readlines();
proper_output = open('{folder}/proper.txt'.format(folder=folder),           'r').readlines();
salign_output = open('{folder}/salign.txt'.format(folder=folder),           'r').readlines();

for i in range(0, 10):
    if term == 'mf':
        term_full = 'Molecular Functionality'
        req_col = 2
        label_y = "Average Functional Similarity"
        color = 'r'
    elif term == 'bp':
        term_full = 'Biological Process'
        req_col = 3
        label_y = 'Average Functional Similarity'
        color = 'b'            
    else:
        index_c = 3
        color = 'c'
        label_y = 'Cellular Component'
# print(salign_output[1:11])
top_limit = 11
salign      = required_columns(salign_output[1:top_limit],      req_col)
modulealign = required_columns(modulealign_output[1:top_limit], req_col)
hubalign    = required_columns(hubalign_output[1:top_limit],    req_col)
ibnal       = required_columns(ibnal_output[1:top_limit],       req_col)
netal       = required_columns(netal_output[1:top_limit],       req_col)
proper      = required_columns(proper_output[1:top_limit],      req_col)
magnapp     = required_columns(magnapp_output[1:top_limit],     req_col)

index = np.arange(10)
bar_width = 0.250
opacity = 0.8

patterns = [ '/', 'X', 'x', 'O', '#', '.', '$' ]

plt.figure( figsize=(50, 40))    

plt.bar(index , salign, bar_width,
alpha=opacity,
color='grey',
hatch= patterns[0],
label='SAlign')
plt.tick_params(labelsize = 50)

plt.bar(index+ bar_width, hubalign, bar_width,
alpha=opacity,
color='black', hatch= patterns[1],
label='HubAlign')

plt.bar(index + (2*bar_width), modulealign, bar_width,
alpha=opacity,
color='blue',hatch= patterns[2],
label='ModuleAlign')
plt.tick_params(labelsize = 50)

# plt.bar(index + (3*bar_width), ibnal, bar_width,
# alpha=opacity,
# color='green',hatch= patterns[3],
# label='IBNAL')
# plt.tick_params(labelsize = 50)

# plt.bar(index + (4*bar_width), netal, bar_width,
# alpha=opacity,
# color='yellow',hatch= patterns[4],
# label='NETAL')
# plt.tick_params(labelsize = 50)

# plt.bar(index + (5*bar_width), proper, bar_width,
# alpha=opacity,
# color='red',hatch= patterns[6],
# label='PROPER')
# plt.tick_params(labelsize = 50)

# plt.bar(index + (6*bar_width), magnapp, bar_width,
# alpha=opacity,
# color='purple',hatch= patterns[5],
# label='MAGNA++')
# plt.tick_params(labelsize = 50)

# plt.ylabel(label_y, fontsize="70")
# plt.title( ' For '+ term_full, fontsize="70")
# plt.legend(prop={'size': 40})

plt.savefig("/Users/imran/Documents/semantic/output/v1/ready-to-plot/{term}".format(term=term))
plt.show()
```
