```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

def required_columns(req_arr, index_c):
    formatted_arr = []
    for req in req_arr:
        col = req.split()
        formatted_arr.append(float(col[index_c]))
    return formatted_arr

term = 'mf'
graph_criteria = 'Scores'
for threshold in range(1, 10):
    folder = "/Users/imran/Documents/semantic/output/v1/threshold/{threshold}".format(threshold=threshold)
    if graph_criteria == 'nodes':
        req_col = 3
        criteria_count= "Nodes count"
    elif graph_criteria == 'scores':
        req_col = 2
        criteria_count = "Scores"
    
    hubalign_output = open('{folder}/hubalign.txt'.format(folder=folder),       'r').readlines();
    ibnal_output = open('{folder}/ibnal.txt'.format(folder=folder),             'r').readlines();
    magnapp_output = open('{folder}/magnapp.txt'.format(folder=folder),         'r').readlines();
    modulealign_output = open('{folder}/modulealign.txt'.format(folder=folder), 'r').readlines();
    netal_output = open('{folder}/netal.txt'.format(folder=folder),             'r').readlines();
    proper_output = open('{folder}/proper.txt'.format(folder=folder),           'r').readlines();
    salign_output = open('{folder}/salign.txt'.format(folder=folder),           'r').readlines();

    for i in range(0, 10):
        if term == 'mf':
            term_full = 'Molecular Functionality'
            index_c = 1
            label_y = "Average Functional Similarity"
            color = 'r'
        elif term == 'bp':
            term_full = 'Biological Process'
            index_c = 2
            label_y = 'Average Functional Similarity'
            color = 'b'            
        else:
            index_c = 3
            color = 'c'
            label_y = 'Cellular Component'

    salign      = required_columns(salign_output[index_c::3],      req_col)
    modulealign = required_columns(modulealign_output[index_c::3], req_col)
    hubalign    = required_columns(hubalign_output[index_c::3],    req_col)
    ibnal       = required_columns(ibnal_output[index_c::3],       req_col)
    netal       = required_columns(netal_output[index_c::3],       req_col)
    proper      = required_columns(proper_output[index_c::3],      req_col)
    magnapp     = required_columns(magnapp_output[index_c::3],     req_col)

    index = np.arange(10)
    bar_width = 0.10
    opacity = 0.8

    patterns = [ '/', 'X', 'x', 'O', '#', '.', '$' ]

    plt.figure( figsize=(50, 40))    

    plt.bar(index , salign, bar_width,
    alpha=opacity,
    color='grey',
    hatch= patterns[0],
    label='SAlign')
    plt.tick_params(labelsize = 50)

    plt.bar(index+ bar_width, hubalign, bar_width,
    alpha=opacity,
    color='black', hatch= patterns[1],
    label='HubAlign')

    plt.bar(index + (2*bar_width), modulealign, bar_width,
    alpha=opacity,
    color='blue',hatch= patterns[2],
    label='ModuleAlign')
    plt.tick_params(labelsize = 50)

    plt.bar(index + (3*bar_width), ibnal, bar_width,
    alpha=opacity,
    color='green',hatch= patterns[3],
    label='IBNAL')
    plt.tick_params(labelsize = 50)

    plt.bar(index + (4*bar_width), netal, bar_width,
    alpha=opacity,
    color='yellow',hatch= patterns[4],
    label='NETAL')
    plt.tick_params(labelsize = 50)

    plt.bar(index + (5*bar_width), proper, bar_width,
    alpha=opacity,
    color='red',hatch= patterns[6],
    label='PROPER')
    plt.tick_params(labelsize = 50)

    plt.bar(index + (6*bar_width), magnapp, bar_width,
    alpha=opacity,
    color='purple',hatch= patterns[5],
    label='MAGNA++')
    plt.tick_params(labelsize = 50)

    plt.ylabel(label_y, fontsize="70")
    plt.title(criteria_count + ' For '+ term_full + " at threshold " + str(float(threshold)/10.0), fontsize="70")
    plt.legend(prop={'size': 40})

    plt.savefig("/Users/imran/Documents/semantic/output/v1/score-based/{threshold}_{term}"
                .format(threshold=threshold, term=term))
    plt.show()
```
