```python
import matplotlib.pyplot as plt
import numpy as np

file = '/Users/imran/Documents/semantic/output/bma.txt'
rs = open(file, 'r').readlines()

def plot_graph(start, end):
    first_121_rows = rs[start: end]
    maxs = []

    sp = []
    term = 'nodes'
    if term == 'nodes':
        label_y = "Connected Nodes"
        index_c = 5
        color = 'r'
    elif term == 'edges':
        label_y = 'LCCS'
        color = 'b'
        index_c = 6
    elif term == 'ec':
        label_y = 'Edge Correctness'
        color = 'b'
        index_c = 11
    elif term == 'sss':
        color = 'r'
        label_y = 'S3'
        index_c = 12   

    for r in first_121_rows:
        col = r.split()
        sp.append(float(col[index_c]))

    maxs.append( [sp.index(max(sp)), max(sp)])
    maxs.append([sp.index(min(sp)), min(sp)])
    maxs.append([sp.index(sp[36]), sp[36]])

    label_line = ''
    label_line = 'iteration \t nodes \t edges \t\t EC \t\t\t s3 \n'
    for mx in maxs:
        max_val_row = ''
        iterr = mx[0]
        max_val_row = first_121_rows[iterr].split()

        al = max_val_row[5]
        b = max_val_row[6]
        sq = max_val_row[11]
        strr = max_val_row[12]
        if iterr < 10:
            iterr = "0{iterr}".format(iterr=iterr)
        label_line = label_line + '{iterr}   \t\t {al} \t {b} \t\t {sq} \t\t\t {strr} \n'.format(maxx=mx[1], iterr=iterr, al=al, b=b, strr=strr, sq=sq)

    title=first_121_rows[1].split()[1]
    plt.figure( figsize=(50, 15))    
    plt.xticks(np.arange(0,122,1))
    plt.plot(sp,'{color}^'.format(color=color), linestyle='solid', label=label_line.expandtabs())
    plt.title(title, fontsize=50)
    plt.xlabel('Iterations ->', fontsize=30)
    plt.ylabel("Nodes", fontsize=30)
    lgnd = plt.legend(prop={'size': 25}, bbox_to_anchor = (0.5, -0.3), loc = 'center')
    plt.savefig("plots/bma/legend_display_individual_pair/topological/{term}/{title}_{term}".format(title=title, term=term), bbox_extra_artists=[lgnd], bbox_inches='tight')

for i in range(0, 9):
    if (i == 0):
        prev_end = 0
        i = 1
    start = 1 + prev_end
    end = start + 121
    prev_end = end
    
    plot_graph(start, end)
    
```
