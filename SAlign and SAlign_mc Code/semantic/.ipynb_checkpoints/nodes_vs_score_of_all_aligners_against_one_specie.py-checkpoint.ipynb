```python
import matplotlib as plt
import numpy as np
from pylab import *
import pandas as pd
#import m

def required_columns(req_arr, index_c):
    formatted_arr = []
    for req in req_arr:
        col = req.split()
        formatted_arr.append(float(col[index_c]))
    return formatted_arr

sp_arr = ['mouse-human','yeast-human','mouse-yeast','mouse-wormm','mouse-flyyy','wormm-flyyy','wormm-human','flyyy-human','yeast-flyyy','wormm-yeast']
sp_arr = ['wormm-yeast']
org_length = 4486
term = 'mf'
for sp in sp_arr:
    for threshold in range(0, 1):
        folder = "/Users/imran/Documents/semantic/mf-threshold.xlsx"

        output = pd.read_excel(folder, index_col=None, header=None);
        data = output.loc[output[0] == sp]

        for i in range(0, 10):
            if term == 'mf':
                term_full = 'Molecular Functionality'
                index_c = 1
                label_y = "Average Functional Similarity"
                color = 'r'
            elif term == 'bp':
                term_full = 'Biological Process'
                index_c = 2
                label_y = 'Average Functional Similarity'
                color = 'b'            
            else:
                index_c = 3
                color = 'c'
                label_y = 'Cellular Component'
        salign      = [round((float(item)/org_length)*100, 2) for item in data[2].tolist()]
        modulealign = [round((float(item)/org_length)*100, 2) for item in data[4].tolist()]
        hubalign    = [round((float(item)/org_length)*100, 2) for item in data[6].tolist()]
        ibnal       = [round((float(item)/org_length)*100, 2) for item in data[8].tolist()]
        netal       = [round((float(item)/org_length)*100, 2) for item in data[10].tolist()]
        magnapp     = [round((float(item)/org_length)*100, 2) for item in data[12].tolist()]
        proper      = [round((float(item)/org_length)*100, 2) for item in data[14].tolist()]

        bar_width = 0.10
        opacity = 0.8

        patterns = [ '/', 'X', 'x', 'O', '#', '.', '$' ]
        fig,ax = plt.subplots(figsize = (10,5))

        plt.plot(salign, '{color}^'.format(color='r'), linestyle='solid', label='SAlign')
        plt.plot(modulealign, '{color}^'.format(color='g'), linestyle='solid', label='ModuleAlign')
        plt.plot(hubalign, '{color}^'.format(color='b'), linestyle='solid', label='HubAlign')
        plt.plot(ibnal, '{color}^'.format(color='y'), linestyle='solid', label='IBNAL')
        plt.plot(netal, '{color}^'.format(color='m'), linestyle='solid', label='NETAL')
        plt.plot(magnapp, '{color}^'.format(color='c'), linestyle='solid', label='MAGNA++')
        plt.plot(proper, '{color}^'.format(color='k'), linestyle='solid', label='PROPER')
        plt.xlabel(term.upper() + " similarity threshold values for "+ sp + " Pair")
        plt.ylabel("percentage of aligned nodes")
        x = np.arange(0,1.0,0.1)
        for i in range(len(x)):
            x[i] = round(x[i],2)
            x[i] = str(x[i])
        
        ax.set_xticklabels(x)
        plt.legend(prop={'size': 10})
        plt.savefig("/Users/imran/Documents/semantic/output/v2/{term}/plots/per/{sp}"
                    .format(sp=sp, term=term))
        plt.show()
```
